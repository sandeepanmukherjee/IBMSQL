```python
census_data = pd.read_csv("CENSUS_DATA.csv")
chicago_schools = pd.read_csv("CHICAGO_PUBLIC_SCHOOLS.csv")
chicago_crime = pd.read_csv("CHICAGO_CRIME_DATA.csv")
```

### Problem 4: List all kidnapping crimes involving a child?

### Problem 5: What kind of crimes were recorded at schools?

```python
schools_crime = chicago_crime[chicago_crime["LOCATION_DESCRIPTION"].str.contains("SCHOOL")]
schools_crime
```

```python
schools_crime.groupby("PRIMARY_TYPE").count()
```
